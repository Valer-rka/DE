```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df = pd.DataFrame(['A', 'B', 'B', 'C'], columns=['Col']) 
df.head()
```

```python
df['Col2']=['K', 'S', 'V', 'N']
df
```

```python
pd.factorize(df['Col'])[1]
```

```python
df['Fact'] = pd.factorize(df['Col'])[0]
df.head()
```

```python
le = LabelEncoder()
df['Lab'] = le.fit_transform(df['Col'])
df.head()
```

```python
df2 = df.copy()
```

```python
df2 = pd.get_dummies(df2)
df2.head()
```

```python
enc = OneHotEncoder()
train=enc.fit_transform(df2)
df3 = pd.DataFrame(train.toarray())
df3.head()
```

```python
enc = OneHotEncoder(sparse=False)
train=enc.fit_transform(df2)
df3 = pd.DataFrame(train)
df3.head()
```

```python
enc = OneHotEncoder(sparse=False)
train=enc.fit_transform(df2[['Col','Col2']])
enc_col = enc.get_feature_names(['Col','Col2'])
df3 = pd.DataFrame(train, columns=enc_col)
df3.head()
```

```python
enc = OneHotEncoder(sparse=False)
train=enc.fit_transform(df2[['Col']])
df3 = pd.DataFrame(train, columns=enc.categories_)
df3.head()
```

```python
rez=pd.concat([df,df3],axis=1)
```

```python
rez
```

```python
rez.drop(['Col'],axis=1, inplace=True)
```
